### Adding Location, Latitude and Longitude

Following the code here to get latitude and longitude: 
https://www.kaggle.com/ryanglasnapp/quick-look-at-sakura-data/notebook

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
test = first_bloom.loc[0,'Site Name']

locator = Nominatim(user_agent="asjew")
location = locator.geocode(test)


print(location.address)
print(f"Latitude = {location.latitude}, Longitude = {location.longitude}")
```

```python
first_bloom.loc[:,'Site Name'] = first_bloom.loc[:,'Site Name'] + ', Japan'
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
first_bloom['Location'] = first_bloom['Site Name'].apply(geocode)
first_bloom['Latitude'] = first_bloom['Location'].apply(lambda x: x.latitude if x else None)
first_bloom['Longitude'] = first_bloom['Location'].apply(lambda x: x.longitude if x else None)
```

```python
# Now for the full bloom data

full_bloom.loc[:,'Site Name'] = full_bloom.loc[:,'Site Name'] + ', Japan'
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
full_bloom['Location'] = full_bloom['Site Name'].apply(geocode)
full_bloom['Latitude'] = full_bloom['Location'].apply(lambda x: x.latitude if x else None)
full_bloom['Longitude'] = full_bloom['Location'].apply(lambda x: x.longitude if x else None)
```

```python
# Now add columns to csv files
first_bloom.to_csv('sakura_first_bloom_dates.csv', index=False)
full_bloom.to_csv('sakura_full_bloom_dates.csv', index=False)
```
